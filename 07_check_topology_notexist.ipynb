```python
import pickle
import glob
import numpy as np
import pandas as pd
```

```python
summ = []

home = '/home/platform/project/crustdb_platform/crustdb_api/workspace/crustdb_database/topology/'
species_list = ['Axolotls', 'Lung']
for species in species_list:
    # print(species, end=' ')
    _conformation_list = glob.glob(home+species+'/*')
    conformation_list = [i.split('/')[-1] for i in _conformation_list]
    for conf in conformation_list:
        # print(conf, end=' ')
        _graph_list = glob.glob(home+species+'/'+conf+'/*')
        graph_list = sorted([i.split('/')[-1] for i in _graph_list])
        if 'corr.csv' not in graph_list:
            continue
        graph_list.remove('corr.csv')
        for g in graph_list:
            # print(g, end=' ')
            _graph_detail_list = glob.glob(home+species+'/'+conf+'/'+g+'/*')
            graph_detail_list = [i.split('/')[-1] for i in _graph_detail_list]
            if 'node.csv' not in graph_detail_list or 'graph.csv' not in graph_detail_list or 'edge.csv' not in graph_detail_list:
                continue
            graph_detail_list.remove('node.csv')
            graph_detail_list.remove('graph.csv')
            graph_detail_list.remove('edge.csv')
            if 'MST_2D.png' in graph_detail_list: graph_detail_list.remove('MST_2D.png')
            if 'MST_3D.png' in graph_detail_list: graph_detail_list.remove('MST_3D.png')
            graph_detail_list = [i.split('.pkl')[0] for i in graph_detail_list]
            # print(graph_detail_list)
            # for g_detail in graph_detail_list:
                # print(g_detail)
            summ.append([species, conf, g, graph_detail_list])
    #             break
    #         break
    #     break
    # break                
```

```python
df = pd.DataFrame(summ, columns=['folder', 'conformation', 'graph_type', 'pkl'])
df
```
